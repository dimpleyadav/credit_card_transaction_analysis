```python
import pandas as pd
from datetime import datetime

# read the csv file
df = pd.read_csv('C:\\Users\\HP\\Desktop\\Datasets\\financial_credit_card_transactions.csv')
# display the first few rows
df.head()
```

```python
# summary statistics
df.describe()
```

### 1. What are the total transaction amounts for each year?

```python
# extract year from transaction date
date = pd.DatetimeIndex(df['Transaction Date'])
df['Year'] = date.year
df.head()
```

```python
# group the dataframe by the 'Year' column and then sum the 'Transaction Amount' for each year
df.groupby('Year')['Transaction Amount'].sum().reset_index()
```

### 2. What are the minimum and maximum transaction amounts in the dataset?

```python
# return the rows where the 'Transaction Amount' is equal to the minimum value found in that column 
df.loc[df['Transaction Amount'] == df['Transaction Amount'].min()]
```

```python
# return the rows where the 'Transaction Amount' is equal to the maximum value found in that column
df.loc[df['Transaction Amount'] == df['Transaction Amount'].max()]
```

### 3. Which merchant categories have the highest average transaction amounts?

```python
# average transaction amount by merchant category
grouped_val = df.groupby('Merchant Category')['Transaction Amount'].mean().reset_index()
grouped_val.sort_values(by = 'Transaction Amount', ascending = False)
```

### 4. What is the total transaction amount by customer gender?

```python
# total transaction amounts by customer gender
df.groupby('Customer Gender')['Transaction Amount'].sum().reset_index()
```

### 5. How many customers have a credit score above 750 ?

```python
# count of high credit score customers
count = (df['Credit Score'] > 750).sum()
count
```

### 6. How does the average transaction amount vary across different age groups?

```python
# define age groups
age_bins = [18, 25, 35, 45, 55, 65, 70]
age_labels = ['18-25', '26-35', '36-45', '46-55', '56-65', '66-70']
df['Age Group'] = pd.cut(df['Customer Age'], bins=age_bins, labels=age_labels, right=False)

# calculate the average transaction amount for each age group
average_transaction_by_age_group = df.groupby('Age Group')['Transaction Amount'].mean().reset_index()

# rename columns for clarity
average_transaction_by_age_group.columns = ['Age Group', 'Average Transaction Amount']

average_transaction_by_age_group.sort_values(by ='Average Transaction Amount', ascending = False)
```

### 7. Is there a significant linear relationship between Customer Income and Transaction Amount?

```python
# Calculate correlation between column 'A' and 'B'
corr_matrix = df['Transaction Amount'].corr(df['Customer Income'])
corr_matrix
```

The correlation coefficient indicates a very weak negative linear relationship between Transaction Amount and Customer Income. This suggests that income is not a strong predictor of transaction amounts in the dataset.
